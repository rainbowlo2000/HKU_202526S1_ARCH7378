```python
import pandas as pd

# prerequisites
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['font.family'] = 'serif'
import numpy as np

import torch
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
to_tensor = lambda x: torch.tensor(x).to(torch.float32)
```

```python
fp = '../../../Datasets/3var_7bar_truss_26x26x26/data.csv'

df = pd.read_csv(fp)

input_columns = ['in:dy_4', 'in:dx_0', 'in:dy_0']  # Adjust as needed
output_column = 'out:Mass_CSOptim'  # Adjust as needed
mask_column = 'out:HasFailed_CSOptim'

X_0, y_0 =  to_tensor(df[input_columns].values).to(torch.float32), to_tensor(df[output_column].values).to(torch.float32).reshape(-1,1)
mask = to_tensor(df[mask_column].values).to(torch.bool)

X = X_0.clone()
y = y_0.clone()

# rescaling really big values
rescale_p = 2
if rescale_p > 1:
    y = y ** (1/rescale_p)
```

```python
X_min, X_max = X_0.min(dim=0)[0], X_0.max(dim=0)[0]
X_range = torch.stack((X_min, X_max), dim=1)
y_min, y_max = y.min(), y.max()
y_min, y_max
X.shape, y.shape, X[:, 0].shape
y_min, y_max

X.shape, y.shape
```

```python


def plot_pairwise_with_colormap(X, y, y_range=None, cmap='viridis'):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    color_props = dict()
    color_props.setdefault('cmap', cmap)
    if y_range:
        color_props.update(dict(vmin=y_range[0], vmax=y_range[1]))
    
    # Create a grid for heatmap background
    for i, ax in enumerate(axs):
        ax.scatter(X[:, i], X[:, (i+1)% 3], s=100, c=y.flatten(), marker='s', **color_props)
        ax.set_xlabel(f'X{i}')
        ax.set_ylabel(f'X{(i+1)%3}')
        ax.grid(True)
        ax.set_xlim(*X_range[i].tolist())
        ax.set_ylim(*X_range[(i+1) % 3].tolist())

    plt.tight_layout()
    plt.show()

plot_pairwise_with_colormap(X, y, [y_min, y_max])
```

```python
# scale data to observed value range
    
p_test = 0.1

scaler_X = MinMaxScaler().fit(X)
scaler_y = MinMaxScaler().fit(y)

X_scaled = to_tensor(scaler_X.transform(X))
y_scaled = to_tensor(scaler_y.transform(y))

# Assuming you have your data tensor `X` and corresponding labels `y`
# X should be of shape (num_samples, num_features)
# y should be of shape (num_samples,)
# Split data into train and test sets
X_train_scaled, X_test_scaled, y_train_scaled, y_test_scaled = train_test_split(X_scaled, y_scaled, test_size=p_test, random_state=42)

# Create TensorDatasets for train and test data
train_dataset = TensorDataset(X_train_scaled, y_train_scaled)
test_dataset = TensorDataset(X_test_scaled, y_test_scaled)

# Create DataLoader instances for train and test datasets
batch_size = 24  # Choose your desired batch size
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

unscale_x = lambda x: to_tensor(scaler_x.inverse_transform(x))
unscale_y = lambda y: to_tensor(scaler_y.inverse_transform(y)) ** (rescale_p)
```

```python
input_size = len(input_columns)
output_size = 1

# Neural Network Model
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_size, 64)  # Input layer to hidden layer
        self.fc2 = nn.Linear(64, 32)  # Hidden layer to hidden layer
        self.fc3 = nn.Linear(32, output_size)  # Hidden layer to output layer

    def forward(self, x):
        x = torch.nn.functional.leaky_relu(self.fc1(x))
        x = torch.nn.functional.leaky_relu(self.fc2(x))
        return self.fc3(x)

def train_model(model, criterion, optimizer, train_loader, losses, epochs=1000):
    model.train()
    for epoch in range(epochs):
        loss_e, loss_unscaled_e = 0., 0.
        for iter, (inputs, targets) in enumerate(train_loader):
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

            loss_unscaled = SMAPELoss(unscale_y(outputs.detach()), unscale_y(targets))

            loss_e += loss.detach()
            loss_unscaled_e += loss_unscaled.detach()
            
        loss_e /= iter
        loss_unscaled_e /= iter
        losses['train'] += [loss_e]
        losses['e_train'] += [epoch]
        if (epoch+1) % 10 == 0:
            print(f'TRAIN | Epoch {epoch+1}, Loss: {loss_e.item():.4f}, SMAPE_unscaled: {loss_unscaled_e.item():.4f}')
            test(model, epoch, losses)

def test(model, epoch, losses):
    _ = model.eval()
    with torch.no_grad():
        loss_e, loss_unscaled_e = 0., 0.
        for iter, (inputs, targets) in enumerate(test_loader):
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss_unscaled = SMAPELoss(unscale_y(outputs.detach()), unscale_y(targets))
            
            loss_e += loss.detach()
            loss_unscaled_e += loss_unscaled.detach()

        loss_e /= iter
        loss_unscaled_e /= iter
        losses['test'] += [loss_e]
        losses['e_test'] += [epoch]
            
    print(f'... TEST | Epoch {epoch+1}, Loss: {loss_e.item():.4f}, SMAPE_unscaled: {loss_unscaled_e.item():.4f}')
```

```python

def SMAPELoss(output, target):
    denominator = (torch.abs(target) + torch.abs(output)) / 2
    diff = torch.abs(output - target) / denominator
    diff[denominator == 0] = 0.0  # Handle the case to avoid division by zero
    smape = torch.mean(diff)
    return smape

# criterion = SMAPELoss
criterion = nn.MSELoss()
```

```python
# Train model

model = Net()

learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

losses = {'train':[], 'test':[], 'e_train':[], 'e_test':[]}
train_model(model, criterion, optimizer, train_loader, losses,  epochs=200)
```

```python
import matplotlib.pyplot as plt

# Plotting the loss
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(losses['e_train'], losses['train'], label='Train Loss', c='blue')
ax.plot(losses['e_test'], losses['test'], label='Test Loss', c='red')

# Adding labels and title
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.set_title('Training and Test Loss')

# Adding legend
ax.legend()

# Displaying the plot
plt.show()
```
